```python
from IPython.display import clear_output

play = True

def replay():   
    ans = input("Do you want to play again Yes/No:")   
    if ans.lower() == 'yes':
        return True
    else:
        return False

while play:
    clear_output()
    test_board = [' ']*10
    
    #to display board
    def display_board(board):
        clear_output()
        print (" "+board[7]+" | "+board[8]+" | "+board[9])
        print ('---|---|---')
        print (" "+board[4]+" | "+board[5]+" | "+board[6])
        print ('---|---|---')
        print (" "+board[1]+" | "+board[2]+" | "+board[3])
        
#     display_board(test_board)
    
    #to get initial input from user
    def player_input():
        marker = ""
        while marker!='X' and marker!='O':
            marker = input("Player 1 will you be 'X' or 'O': ")
        
        if marker == 'X':
            player2 = 'O'
        else:
            player2 = 'X'
        return (marker,player2)
    
    player1_marker, player2_marker = player_input()
    
    def win_check(board, mark):
    
        checklist = []
        #print("inside win check")
        def diff(list1,list2):
            list3 = list(set(list2) - set(list1))
            return list3
    
        #getting all positions of X from the board
        for x in range(1,10):
            if mark == board[x]:
                checklist.append(x)
            
        #possible patterns for winnning
        newlist = [[1,2,3],[1,4,7],[1,5,9],[2,5,8],[3,5,7],[3,6,9],[4,5,6],[7,8,9]]
    
        for item_list in newlist:
            result = diff(checklist,item_list)
            if not result:
                return "Won"
            else:
                pass
        return "lost"
    
    def place_marker(board, marker, position):
        board[position] = marker
        display_board(board)
        result = win_check(board,marker)
        #print(result)
        if result == "Won":
            #print("inside place_marker if")
            return True #pass this to player choice
        else:
            return False
         
    def space_check(board, position):
        if board[position] == ' ':
            return True
        else:
            return False
    
    #if board is full it will return True and if some space is there it will return False
    def full_board_check(board):
        for i in range(1,10):
            if board[i] != ' ':
                continue
            else:
                return False
        return True
    
    def player_choice(board,marker):
        player_position = 0
        while (1 > player_position or player_position > 9):
            player_position = int(input("Give your next move:"))
    
        #check if there is space at the asked position
        result = space_check(board,player_position)
        if result:
            ans = place_marker(board,marker,player_position)
            #print(ans)
            if ans:
                return True #pass this to player call
            else:
                return False
        else:
            print("please make another choice")
            player_choice(board,marker)
            

    game_on = True
    
    while game_on:
        
        def player1():
            if not full_board_check(test_board):
                choice1 = player_choice(test_board,player1_marker)
                if choice1:
                    return False #pass this to game on
                else:
                    return True
            else:
                print("Board Full")
                return False
                #game_on = False
        
        def player2():
            if not full_board_check(test_board):
                choice2 = player_choice(test_board,player2_marker)
                if choice2:
                    return False
                else:
                    return True           
            else:
                print("Board Full")
                return False
    
        game_on = player1()
        if game_on == False:
            break
        game_on = player2()
        if game_on == False:
            break
            
    play = replay()

print("Thank you for playing this game")
```
