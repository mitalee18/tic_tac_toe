```python
from IPython.display import clear_output

play = True

def replay():   
    ans = input("Do you want to play again Yes/No:").lower() 
    if ans == 'yes':
        return True
    else:
        return False

while play:
    test_board = [' ']*10
    
    #to display board
    def display_board(board):
        clear_output()
        print (" "+board[7]+" | "+board[8]+" | "+board[9])
        print ('---|---|---')
        print (" "+board[4]+" | "+board[5]+" | "+board[6])
        print ('---|---|---')
        print (" "+board[1]+" | "+board[2]+" | "+board[3])
    
    #to get initial input from user
    def player_input():
        marker = ""
        while marker!='X' and marker!='O':
            marker = input("Player 1 will you be 'X' or 'O': ").upper()
        player2 = 'X' if marker == 'O' else 'O'
        return (marker,player2)
    
    player1_marker, player2_marker = player_input()
    
    def win_check(board, mark):
    
        checklist = []
        #print("inside win check")
        def diff(list1,list2):
            list3 = list(set(list2) - set(list1))
            return list3
    
        #getting all positions of X from the board
        for x in range(1,10):
            if mark == board[x]:
                checklist.append(x)
            
        #possible patterns for winnning
        newlist = [[1,2,3],[1,4,7],[1,5,9],[2,5,8],[3,5,7],[3,6,9],[4,5,6],[7,8,9]]
    
        for item_list in newlist:
            if not diff(checklist,item_list):
                return True #true for win return to placemarker
            else:
                pass
        return False
    
    def place_marker(board, marker, position):
        board[position] = marker
        display_board(board)
        #returns true if win, and then return that true to player_choice
        return win_check(board,marker)
         
    def space_check(board, position):
        return board[position] == ' '
    
    #if board is full it will return True and if some space is there it will return False
    def full_board_check(board):
        return ' ' not in board[1:]  
    
    def player_choice(board,marker):
        player_position = 0
        while player_position not in range(1,10):
            player_position = int(input("Give your next move player1:")) if marker == player1_marker else int(input("Give your next move player2:"))
    
        #check if there is space at the asked position
        if space_check(board,player_position):
            return place_marker(board,marker,player_position)
        else:
            print("Position already full, please make another choice")
            player_choice(board,marker)

    game_on = True
    
    while game_on:
        
        def player1():
            if not full_board_check(test_board):
                if player_choice(test_board,player1_marker):
                    return False #pass this to game on
                else:
                    return True
            else:
                print("It's a draw. The board is Full")
                return False
        
        def player2():
            if not full_board_check(test_board):
                if player_choice(test_board,player2_marker):
                    return False
                else:
                    return True           
            else:
                print("It's a draw. The board is Full")
                return False
    
        game_on = player1()
        if game_on == False:
            print("Congratulations Player 1 you win!")
            break
        game_on = player2()
        if game_on == False:
            print("Congratulations Player 2 you win!")
            break
            
    play = replay()

print("Thank you for playing this game")
```
