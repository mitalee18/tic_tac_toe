```python
from IPython.display import clear_output
print('Welcome to Tic Tac Toe!')
    
play = True
#while game_on:
while play:
        
    # Set the game up here
    test_board = [' ']*10
    
    def display_board(board):
    
        print (" "+board[7]+" | "+board[8]+" | "+board[9])
        print ('---|---|---')
        print (" "+board[4]+" | "+board[5]+" | "+board[6])
            print ('---|---|---')
            print (" "+board[1]+" | "+board[2]+" | "+board[3])    
    
        
        #replay
        def replay():
    
            ans = input("Do you want to play again Yes/No:")
            if ans.lower() == 'yes':
                return True
            else:
                return False
        
        #won
        def you_win():
            print("Congratulations you win!")
            play = replay()
        
        #check the board
        def full_board_check(board):
            for i in range(1,10):
                if board[i] != ' ':
                    continue
                else:
                    return False
            return True
    
        #space check
        def space_check(board, position):
            if board[position] == ' ':
                return True
            else:
                return False
        
        #check won or not after placing the mark
        def win_check(board, mark):
            checklist = []
    
            def diff(list1,list2):
                list3 = list(set(list2) - set(list1))
                return list3
    
            #getting all positions of X from the board
            for x in range(1,10):
                if mark == board[x]:
                    checklist.append(x)
            
            #possible patterns for winnning
            newlist = [[1,2,3],[1,4,7],[1,5,9],[2,5,8],[3,5,7],[3,6,9],[4,5,6],[7,8,9]]
    
            for item_list in newlist:
                result = diff(checklist,item_list)
                if not result:
                    return "Won"
                else:
                    pass
            return "lost"


        #setting marker
        def place_marker(board, marker, position):
            if space_check(board,position):
                board[position] = marker
                result = win_check(board,marker)
                if result == "Won":
                    you_win()
                else:
                    display_board(board)
                    return True
            else:
                return False
        
        #Player 1 Turn
        def player1(board,player1_marker,player2_marker):
            print("inside plaer1")
            player_position = 0
            while (1 > player_position or player_position > 9):
                player_position = int(input("Enter the position of "+player1_marker+":"))
    
            #check board before adding the marker
            if full_board_check(board):
                if not place_marker(board,player1_marker,player_position):
                    player1(board,player1_marker)
                else:
                    player2(board,player2_marker,player1_marker)
            else:
                pass
                      
        
        # Player2's turn.
        def player2(board,player2_marker,player1_marker):
            
            player_position = 0
            while (1 > player_position or player_position > 9):
                player_position = int(input("Enter the position of "+player2_marker+":"))
            
            #check board before adding the marker
            if full_board_check(board):
                if not place_marker(board,player2_marker,player_position):
                    player2(board,player2_marker,player1_marker)
                else:
                    player1(board,player1_marker,player2_marker)
            else:
                pass
             
        def player_input():
            marker = ""
            while marker!='X' and marker!='O':
                marker = input("Player 1 will you be 'X' or 'O': ")
        
                if marker == 'X':
                    player2 = 'O'
                else:
                    player2 = 'X'
    
            return (marker,player2)
    
        player1_marker, player2_marker = player_input()
        
        
        def second():
            if player1_marker == 'X':
                print("Woohoo PLayer 1 you go First")
                player1(test_board,player1_marker,player2_marker)
            else:
                print("Woohoo PLayer 2 you go First")
                player2(test_board,player2_marker,player1_marker)
        second()
```
